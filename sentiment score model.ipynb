```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
import time
from openpyxl import Workbook
from imdb import IMDb

# Function to get IMDb ID using IMDbPY
def get_imdb_id(title_name):
    ia = IMDb()
    try:
        search_results = ia.search_movie(title_name)
        if search_results:
            # Assuming the first result is the correct one
            movie = search_results[0]
            ia.update(movie)
            return movie.movieID
        else:
            return None
    except Exception as e:
        print(f"Error fetching IMDb ID: {e}")
        return None

# Function to get IMDb rating using IMDbPY
def get_imdb_rating(imdb_id):
    ia = IMDb()
    try:
        movie = ia.get_movie(imdb_id)
        rating = movie['rating']
        return rating
    except Exception as e:
        print(f"Error fetching IMDb rating: {e}")
        return None

# Function to get genres using IMDbPY
def get_imdb_genres(imdb_id):
    ia = IMDb()
    try:
        movie = ia.get_movie(imdb_id)
        genres = movie.get('genres', [])
        return genres
    except Exception as e:
        print(f"Error fetching IMDb genres: {e}")
        return []

# Function to scrape IMDb reviews
def scrape_imdb_reviews(title_id, max_reviews=150, max_retries=5):
    reviews = []
    pagination_key = None
    retry_attempts = 0
    while len(reviews) < max_reviews:
        try:
            if pagination_key:
                url = f"https://www.imdb.com/title/tt{title_id}/reviews/_ajax?paginationKey={pagination_key}"
            else:
                url = f"https://www.imdb.com/title/tt{title_id}/reviews?ref_=tt_ql_3"

            response = requests.get(url, timeout=10)
            response.raise_for_status()

            soup = BeautifulSoup(response.content, 'html.parser')
            review_containers = soup.find_all('div', class_='text show-more__control')
            if not review_containers:
                break  # No more reviews found

            for container in review_containers:
                if len(reviews) < max_reviews:
                    reviews.append(container.get_text())
                else:
                    break

            load_more_data = soup.find('div', class_='load-more-data')
            if load_more_data and 'data-key' in load_more_data.attrs:
                pagination_key = load_more_data['data-key']
            else:
                break

            retry_attempts = 0  # Reset retry attempts after a successful request

        except (requests.exceptions.RequestException, requests.exceptions.ConnectTimeout) as e:
            retry_attempts += 1
            if retry_attempts > max_retries:
                print(f"Max retries exceeded: {e}")
                break
            else:
                wait_time = 2 ** retry_attempts  # Exponential backoff
                print(f"Retrying in {wait_time} seconds...")
                time.sleep(wait_time)

    return reviews

# Function to save reviews to Excel
def save_reviews_to_excel(reviews, title_name, source):
    wb = Workbook()
    ws = wb.active
    ws.title = f"{source} Reviews"

    # Add title and column header
    #ws.append([f"Reviews for: {title_name}"])
    ws.append(["Review Number", "Review Text"])

    # Add reviews to the worksheet
    for idx, review in enumerate(reviews, 1):
        ws.append([idx, review])

    # Save the workbook
    file_name = f"{title_name.replace(' ', ' ')}_{source}_reviews.xlsx"
    wb.save(file_name)
    #print(f"Reviews saved to {file_name}")

# Function to find content URL on Rotten Tomatoes
def find_content_url(base_url, search_term):
    search_endpoint = '/search'
    params = {'search': search_term}
    response = requests.get(urljoin(base_url, search_endpoint), params=params)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        search_results = soup.find_all('a', class_='unset', attrs={'data-qa': 'info-name'})
        for result in search_results:
            if search_term.lower() in result.text.strip().lower():
                return urljoin(base_url, result['href'])
        print(f"No exact content link found for '{search_term}'")
        return None
    else:
        print(f'Failed to retrieve search results. Status code: {response.status_code}')
        return None

# Function to get ratings from Rotten Tomatoes
def get_ratings(content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        ratings = {}
        media_scorecard = soup.find('div', class_='media-scorecard')
        if media_scorecard:
            critics_score_button = media_scorecard.find('rt-button', attrs={
                "data-mediascorecardmanager": "overlayOpen:click",
                "slot": "criticsScore",
                "theme": "transparent"
            })
            if critics_score_button:
                critics_rt_text = critics_score_button.find('rt-text', attrs={"context": "label", "size": "1.375"})
                if critics_rt_text:
                    ratings['Critics Score'] = critics_rt_text.text.strip()
            audience_score_button = media_scorecard.find('rt-button', attrs={
                "data-mediascorecardmanager": "overlayOpen:click",
                "slot": "audienceScore",
                "theme": "transparent"
            })
            if audience_score_button:
                audience_rt_text = audience_score_button.find('rt-text', attrs={"context": "label", "size": "1.375"})
                if audience_rt_text:
                    ratings['Audience Score'] = audience_rt_text.text.strip()
        return ratings if ratings else {"Ratings": "Not found"}
    return {"Failed to retrieve content"}

# Function to get genres from Rotten Tomatoes
def get_genres(content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        genres = []
        genre_elements = soup.find_all('rt-text', attrs={"slot": "genre"})
        if genre_elements:
            genres = [genre.text.strip() for genre in genre_elements]
        if not genres:
            genre_elements = soup.find_all(class_=lambda x: x and 'genre' in x.lower())
            genres = [genre.text.strip() for genre in genre_elements]
        if not genres:
            genre_text = soup.find(text=lambda text: text and "Genre:" in text)
            if genre_text:
                genres = genre_text.find_next(text=True).strip().split(',')
        genres = list(set(filter(None, genres)))
        return genres if genres else ["Genres not found"]
    return ["Genres not found"]

# Function to navigate to reviews page on Rotten Tomatoes
def navigate_to_reviews(base_url, content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        content_soup = BeautifulSoup(response.content, 'html.parser')
        view_all_reviews_link = content_soup.find('rt-link', href=True, context='label', size='0.875')
        if view_all_reviews_link:
            reviews_url = urljoin(base_url, view_all_reviews_link['href'])
            return reviews_url
        else:
            print('View all reviews link not found on content page.')
            return None
    else:
        print(f'Failed to retrieve content page. Status code: {response.status_code}')
        return None

# Function to collect reviews using Selenium
def collect_reviews_selenium(base_url, reviews_url, max_reviews=150):
    options = webdriver.ChromeOptions()
    options.add_argument('--start-maximized')
    options.add_argument('--disable-blink-features=AutomationControlled')
    options.add_argument('--headless') # Run Chrome in headless mode
    options.add_argument('--no-sandbox') # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage') # Overcome limited resource problems
    driver = webdriver.Chrome(options=options) # Use the correct path to ChromeDriver

    # Print Chrome and ChromeDriver versions
    print("Chrome version:", driver.capabilities['browserVersion'])
    print("ChromeDriver version:", driver.capabilities['chrome']['chromedriverVersion'].split(' ')[0])

    driver.execute_cdp_cmd('Network.setUserAgentOverride', {"userAgent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
    driver.get(reviews_url)
    reviews_collected = set()
    try:
        while len(reviews_collected) < max_reviews:
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'p.review-text')))
            review_elements = driver.find_elements(By.CSS_SELECTOR, 'p.review-text')
            for element in review_elements:
                review_text = element.text.strip()
                reviews_collected.add(review_text)
            if len(reviews_collected) >= max_reviews:
                break
            try:
                load_more_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'rt-button[data-qa="load-more-btn"]')))
                driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
                time.sleep(1)
                try:
                    load_more_button.click()
                except ElementClickInterceptedException:
                    ActionChains(driver).move_to_element(load_more_button).click().perform()
                except:
                    driver.execute_script("arguments[0].click();", load_more_button)
                time.sleep(3)
                new_review_count = len(driver.find_elements(By.CSS_SELECTOR, 'p.review-text'))
                if new_review_count <= len(reviews_collected):
                    print("No new reviews loaded. Stopping.")
                    break
            except TimeoutException:
                print("Load More button not found or not clickable. Stopping.")
                break
            except Exception as e:
                print(f"Error clicking Load More button: {e}")
                break
    except Exception as e:
        print(f'Error collecting reviews: {e}')
    finally:
        driver.quit()
    reviews_list = list(reviews_collected)
    return reviews_list[:max_reviews]

# Main function
search_term = None
def main():
    base_url = 'https://www.rottentomatoes.com'
    global search_term

    # Rotten Tomatoes
    content_url = find_content_url(base_url, search_term)
    if content_url:
        print("FOR ROTTEN TOMATOES")
        #print(f"  content URL for '{search_term}' on Rotten Tomatoes: {content_url}")
        ratings = get_ratings(content_url)
        genres = get_genres(content_url)
        print("    Rotten Tomatoes Ratings:")
        if 'Audience Score' in ratings:
            print(f"      Audience Rating: {ratings['Audience Score']}")
        if 'Critics Score' in ratings:
            print(f"      Critics Score: {ratings['Critics Score']}")
        if 'Audience Rating' not in ratings and 'Critics Score' not in ratings:
            print("Ratings not found")
        print(f"      Rotten Tomatoes Genres: {', '.join(genres)}")
        reviews_url = navigate_to_reviews(base_url, content_url)
        if reviews_url:
            #print(f"Reviews URL on Rotten Tomatoes: {reviews_url}")
            collected_reviews = collect_reviews_selenium(base_url, reviews_url, max_reviews=150)
            #print(f"Total collected reviews for '{search_term}' on Rotten Tomatoes: {len(collected_reviews)}")
            save_reviews_to_excel(collected_reviews, search_term, "RottenTomatoes")
        else:
            print(f"Failed to retrieve reviews URL for '{search_term}' on Rotten Tomatoes")
    else:
        print(f"No exact content link found for '{search_term}' on Rotten Tomatoes")

    # IMDb
    imdb_id = get_imdb_id(search_term)
    if imdb_id:
        print("FOR IMDB")
        print(f"IMDb ID for '{search_term}': {imdb_id}")

        # Fetch IMDb rating
        rating = get_imdb_rating(imdb_id)
        if rating:
            print(f"  IMDb Rating: {rating}")
        else:
            print(f"No IMDb rating found for '{search_term}'")

        # Fetch genres
        genres = get_imdb_genres(imdb_id)
        if genres:
            print(f"  Genres: {', '.join(genres)}")
        else:
            print(f"No genres found for '{search_term}'")

        # Scrape IMDb reviews
        print(f"\nFetching reviews for '{search_term}' from IMDb DONE")
        imdb_reviews = scrape_imdb_reviews(imdb_id, max_reviews=150)  # Adjust max_reviews as needed
        if imdb_reviews:
            #print("\nSaving reviews to Excel...")
            save_reviews_to_excel(imdb_reviews, search_term, "IMDb")
        else:
            print("No IMDb reviews found.")
    else:
        print(f"No IMDb ID found for '{search_term}'")




if __name__ == "__main__":
    global search_term
    search_term = input("Enter the name of the movie, series, or show: ")
    main()
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# Set up your Hugging Face token
import os
os.environ["HUGGINGFACE_TOKEN"] = "" # hugging face api key for llama2

#

# Load the Llama 2 model and tokenizer
model_name = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model_name, token=os.environ["HUGGINGFACE_TOKEN"])
model = AutoModelForCausalLM.from_pretrained(model_name, token=os.environ["HUGGINGFACE_TOKEN"],
                                             device_map="auto",
                                             torch_dtype=torch.float16)

```

```python

import re

def analyze_sentiment(review_text):
    prompt = f"""Analyze the sentiment of the following review for the movie Titanic. Provide only score between -1 and 1, where -1 is extremely negative, 0 is neutral, and 1 is extremely positive.
    SCORE: """
    inputs = tokenizer(prompt, return_tensors="pt", max_length=2048, truncation=True).to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=200)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    # Extract the score using regular expressions
    match = re.search(r"SCORE:\s*(-?\d+(\.\d+)?)", generated_text)
    if match:
        score_str = match.group(1)
        # Convert score to float and then to int if possible
        try:
            score = (float(score_str))  # Convert to float first to handle decimal scores
        except ValueError:
            score = "N/A"  # If conversion fails, return "N/A"
    else:
        score = "N/A"  # In case no score is found
    return score

def generate_perspective(review_text):
    prompt = f"""Summarize the overall opinions and feelings of all users about Titanic based on the following reviews. In third person, incorporating both positive and negative aspects, in a 300-word paragraph. Do not copy the reviews.
    Summary:"""


    inputs = tokenizer(prompt, return_tensors="pt", max_length=3000, truncation=True).to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=300)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    # Remove the prompt part from the generated text if it's included
    perspective = generated_text.split("Summary:")[-1].strip()
    return perspective




# Load the Excel file
df1 = pd.read_excel(f"/content/{search_term}_IMDb_reviews.xlsx")
df2= pd.read_excel(f"/content/{search_term}_RottenTomatoes_reviews.xlsx")

# Concatenate all reviews into one large text
combined_reviews_df1 = " ".join(df1['Review Text'].dropna().tolist())
combined_reviews_df2 = " ".join(df2['Review Text'].dropna().tolist())


# Generate the generalized perspective for all users
generalized_perspective_result_df1 = generate_perspective(combined_reviews_df1)
generalized_perspective_result_df2 = generate_perspective(combined_reviews_df2)

# Print the generalized perspective for all users
# Add sentiment analysis for each review
df1['sentiment_score'] = df1['Review Text'].apply(analyze_sentiment)
df2['sentiment_score'] = df2['Review Text'].apply(analyze_sentiment)




```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
import time
from openpyxl import Workbook
from imdb import IMDb

# Function to get IMDb ID using IMDbPY
def get_imdb_id(title_name):
    ia = IMDb()
    try:
        search_results = ia.search_movie(title_name)
        if search_results:
            # Assuming the first result is the correct one
            movie = search_results[0]
            ia.update(movie)
            return movie.movieID
        else:
            return None
    except Exception as e:
        print(f"Error fetching IMDb ID: {e}")
        return None

# Function to get IMDb rating using IMDbPY
def get_imdb_rating(imdb_id):
    ia = IMDb()
    try:
        movie = ia.get_movie(imdb_id)
        rating = movie['rating']
        return rating
    except Exception as e:
        print(f"Error fetching IMDb rating: {e}")
        return None

# Function to get genres using IMDbPY
def get_imdb_genres(imdb_id):
    ia = IMDb()
    try:
        movie = ia.get_movie(imdb_id)
        genres = movie.get('genres', [])
        return genres
    except Exception as e:
        print(f"Error fetching IMDb genres: {e}")
        return []

# Function to scrape IMDb reviews
def scrape_imdb_reviews(title_id, max_reviews=150, max_retries=5):
    reviews = []
    pagination_key = None
    retry_attempts = 0
    while len(reviews) < max_reviews:
        try:
            if pagination_key:
                url = f"https://www.imdb.com/title/tt{title_id}/reviews/_ajax?paginationKey={pagination_key}"
            else:
                url = f"https://www.imdb.com/title/tt{title_id}/reviews?ref_=tt_ql_3"

            response = requests.get(url, timeout=10)
            response.raise_for_status()

            soup = BeautifulSoup(response.content, 'html.parser')
            review_containers = soup.find_all('div', class_='text show-more__control')
            if not review_containers:
                break  # No more reviews found

            for container in review_containers:
                if len(reviews) < max_reviews:
                    reviews.append(container.get_text())
                else:
                    break

            load_more_data = soup.find('div', class_='load-more-data')
            if load_more_data and 'data-key' in load_more_data.attrs:
                pagination_key = load_more_data['data-key']
            else:
                break

            retry_attempts = 0  # Reset retry attempts after a successful request

        except (requests.exceptions.RequestException, requests.exceptions.ConnectTimeout) as e:
            retry_attempts += 1
            if retry_attempts > max_retries:
                print(f"Max retries exceeded: {e}")
                break
            else:
                wait_time = 2 ** retry_attempts  # Exponential backoff
                print(f"Retrying in {wait_time} seconds...")
                time.sleep(wait_time)

    return reviews

# Function to save reviews to Excel
from docx import Document

def save_reviews_to_excel(reviews, title_name, source):
    # Create a new Word document
    doc = Document()

    # Add a title to the document
    #doc.add_heading(f'Reviews for: {title_name}', level=1)

    # Add source information
    #doc.add_paragraph(f'Source: {source}\n')

    # Add reviews to the document
    for idx, review in enumerate(reviews, 1):
        doc.add_paragraph(f'Review {idx}: {review}')

    # Save the document
    file_name = f"{title_name.replace(' ', ' ')}_{source}_reviews.docx"
    doc.save(file_name)
    print(f"Reviews saved to Word file: {file_name}")


# Function to find content URL on Rotten Tomatoes
def find_content_url(base_url, search_term):
    search_endpoint = '/search'
    params = {'search': search_term}
    response = requests.get(urljoin(base_url, search_endpoint), params=params)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        search_results = soup.find_all('a', class_='unset', attrs={'data-qa': 'info-name'})
        for result in search_results:
            if search_term.lower() in result.text.strip().lower():
                return urljoin(base_url, result['href'])
        print(f"No exact content link found for '{search_term}'")
        return None
    else:
        print(f'Failed to retrieve search results. Status code: {response.status_code}')
        return None

# Function to get ratings from Rotten Tomatoes
def get_ratings(content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        ratings = {}
        media_scorecard = soup.find('div', class_='media-scorecard')
        if media_scorecard:
            critics_score_button = media_scorecard.find('rt-button', attrs={
                "data-mediascorecardmanager": "overlayOpen:click",
                "slot": "criticsScore",
                "theme": "transparent"
            })
            if critics_score_button:
                critics_rt_text = critics_score_button.find('rt-text', attrs={"context": "label", "size": "1.375"})
                if critics_rt_text:
                    ratings['Critics Score'] = critics_rt_text.text.strip()
            audience_score_button = media_scorecard.find('rt-button', attrs={
                "data-mediascorecardmanager": "overlayOpen:click",
                "slot": "audienceScore",
                "theme": "transparent"
            })
            if audience_score_button:
                audience_rt_text = audience_score_button.find('rt-text', attrs={"context": "label", "size": "1.375"})
                if audience_rt_text:
                    ratings['Audience Score'] = audience_rt_text.text.strip()
        return ratings if ratings else {"Ratings": "Not found"}
    return {"Failed to retrieve content"}

# Function to get genres from Rotten Tomatoes
def get_genres(content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        genres = []
        genre_elements = soup.find_all('rt-text', attrs={"slot": "genre"})
        if genre_elements:
            genres = [genre.text.strip() for genre in genre_elements]
        if not genres:
            genre_elements = soup.find_all(class_=lambda x: x and 'genre' in x.lower())
            genres = [genre.text.strip() for genre in genre_elements]
        if not genres:
            genre_text = soup.find(text=lambda text: text and "Genre:" in text)
            if genre_text:
                genres = genre_text.find_next(text=True).strip().split(',')
        genres = list(set(filter(None, genres)))
        return genres if genres else ["Genres not found"]
    return ["Genres not found"]

# Function to navigate to reviews page on Rotten Tomatoes
def navigate_to_reviews(base_url, content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        content_soup = BeautifulSoup(response.content, 'html.parser')
        view_all_reviews_link = content_soup.find('rt-link', href=True, context='label', size='0.875')
        if view_all_reviews_link:
            reviews_url = urljoin(base_url, view_all_reviews_link['href'])
            return reviews_url
        else:
            print('View all reviews link not found on content page.')
            return None
    else:
        print(f'Failed to retrieve content page. Status code: {response.status_code}')
        return None

# Function to collect reviews using Selenium
def collect_reviews_selenium(base_url, reviews_url, max_reviews=100):
    options = webdriver.ChromeOptions()
    options.add_argument('--start-maximized')
    options.add_argument('--disable-blink-features=AutomationControlled')
    options.add_argument('--headless') # Run Chrome in headless mode
    options.add_argument('--no-sandbox') # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage') # Overcome limited resource problems
    driver = webdriver.Chrome(options=options) # Use the correct path to ChromeDriver

    # Print Chrome and ChromeDriver versions
    print("Chrome version:", driver.capabilities['browserVersion'])
    print("ChromeDriver version:", driver.capabilities['chrome']['chromedriverVersion'].split(' ')[0])

    driver.execute_cdp_cmd('Network.setUserAgentOverride', {"userAgent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
    driver.get(reviews_url)
    reviews_collected = set()
    try:
        while len(reviews_collected) < max_reviews:
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'p.review-text')))
            review_elements = driver.find_elements(By.CSS_SELECTOR, 'p.review-text')
            for element in review_elements:
                review_text = element.text.strip()
                reviews_collected.add(review_text)
            if len(reviews_collected) >= max_reviews:
                break
            try:
                load_more_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'rt-button[data-qa="load-more-btn"]')))
                driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
                time.sleep(1)
                try:
                    load_more_button.click()
                except ElementClickInterceptedException:
                    ActionChains(driver).move_to_element(load_more_button).click().perform()
                except:
                    driver.execute_script("arguments[0].click();", load_more_button)
                time.sleep(3)
                new_review_count = len(driver.find_elements(By.CSS_SELECTOR, 'p.review-text'))
                if new_review_count <= len(reviews_collected):
                    print("No new reviews loaded. Stopping.")
                    break
            except TimeoutException:
                print("Load More button not found or not clickable. Stopping.")
                break
            except Exception as e:
                print(f"Error clicking Load More button: {e}")
                break
    except Exception as e:
        print(f'Error collecting reviews: {e}')
    finally:
        driver.quit()
    reviews_list = list(reviews_collected)
    return reviews_list[:max_reviews]

# Main function
search_term = None
def main():
    base_url = 'https://www.rottentomatoes.com'
    global search_term

    # Rotten Tomatoes
    content_url = find_content_url(base_url, search_term)
    if content_url:
        print("FOR ROTTEN TOMATOES")
        #print(f"  content URL for '{search_term}' on Rotten Tomatoes: {content_url}")
        ratings = get_ratings(content_url)
        genres = get_genres(content_url)
        print("    Rotten Tomatoes Ratings:")
        if 'Audience Score' in ratings:
            print(f"      Audience Rating: {ratings['Audience Score']}")
        if 'Critics Score' in ratings:
            print(f"      Critics Score: {ratings['Critics Score']}")
        if 'Audience Rating' not in ratings and 'Critics Score' not in ratings:
            print("Ratings not found")
        print(f"      Rotten Tomatoes Genres: {', '.join(genres)}")
        reviews_url = navigate_to_reviews(base_url, content_url)
        if reviews_url:
            #print(f"Reviews URL on Rotten Tomatoes: {reviews_url}")
            collected_reviews = collect_reviews_selenium(base_url, reviews_url, max_reviews=200)
            #print(f"Total collected reviews for '{search_term}' on Rotten Tomatoes: {len(collected_reviews)}")
            save_reviews_to_excel(collected_reviews, search_term, "RottenTomatoes")
        else:
            print(f"Failed to retrieve reviews URL for '{search_term}' on Rotten Tomatoes")
    else:
        print(f"No exact content link found for '{search_term}' on Rotten Tomatoes")

    # IMDb
    imdb_id = get_imdb_id(search_term)
    if imdb_id:
        print("FOR IMDB")
        print(f"IMDb ID for '{search_term}': {imdb_id}")

        # Fetch IMDb rating
        rating = get_imdb_rating(imdb_id)
        if rating:
            print(f"  IMDb Rating: {rating}")
        else:
            print(f"No IMDb rating found for '{search_term}'")

        # Fetch genres
        genres = get_imdb_genres(imdb_id)
        if genres:
            print(f"  Genres: {', '.join(genres)}")
        else:
            print(f"No genres found for '{search_term}'")

        # Scrape IMDb reviews
        print(f"\nFetching reviews for '{search_term}' from IMDb DONE")
        imdb_reviews = scrape_imdb_reviews(imdb_id, max_reviews=100)  # Adjust max_reviews as needed
        if imdb_reviews:
            #print("\nSaving reviews to Excel...")
            save_reviews_to_excel(imdb_reviews, search_term, "IMDb")
        else:
            print("No IMDb reviews found.")
    else:
        print(f"No IMDb ID found for '{search_term}'")




if __name__ == "__main__":
    global search_term
    search_term = input("Enter the name of the movie, series, or show: ")
    main()
```

```python
import time
import re
from docx import Document

# Function to extract text from a Word document
def extract_text_from_word(file_path):
    doc = Document(file_path)
    return [paragraph.text for paragraph in doc.paragraphs if paragraph.text.strip()]

# Load reviews from IMDb Word document
imdb_source_file = f"/content/{search_term}_IMDb_reviews.docx"
text_array_imdb = extract_text_from_word(imdb_source_file)

# Load reviews from Rotten Tomatoes Word document
rotten_tomatoes_source_file = f"/content/{search_term}_RottenTomatoes_reviews.docx"
text_array_rotten_tomatoes = extract_text_from_word(rotten_tomatoes_source_file)

# Function to process reviews in batches
def process_reviews_in_batches(reviews, batch_size=20, source_name='Reviews'):
    num_reviews = len(reviews)
    mean_scores = []

    for i in range(0, num_reviews, batch_size):
        batch_reviews = reviews[i:i + batch_size]

        # Create the prompt for the current batch
        prompt = f"You are tasked with analyzing the polarity of the following reviews. Assign a polarity score to each review on a scale from -1 to 1, where -1 indicates extremely negative sentiment, 0 indicates neutral sentiment, and 1 indicates extremely positive sentiment. Please calculate the mean sentiment score for all the reviews provided. Do not provide individual scores for each review; only return the mean score. {batch_reviews}"

        # Generate the sentiment score using your model
        macro_info = model.generate_content(prompt)

        # Use regex to extract the mean score from the model's output
        match = re.search(r'[-+]?\d*\.\d+|\d+', macro_info.text.strip())
        if match:
            mean_score = float(match.group(0))  # Convert the captured string to float
            mean_scores.append(mean_score)

            # Print the mean score for the current batch
            print(f"Mean Sentiment Score for {source_name} reviews {i+1} to {min(i + batch_size, num_reviews)}: {mean_score}")
        else:
            print(f"Failed to extract mean score from the model's output: '{macro_info.text.strip()}'")

        # Wait for 10 seconds before processing the next batch
        time.sleep(20)

    # Calculate the overall mean score across all batches
    overall_mean_score = sum(mean_scores) / len(mean_scores) if mean_scores else 0
    print(f"Overall Mean Sentiment Score for {source_name}: {overall_mean_score}")

# Process the IMDb reviews in batches
process_reviews_in_batches(text_array_imdb, source_name='IMDb')

print("##--##--##--##--##--##--##--##--##")

# Process the Rotten Tomatoes reviews in batches
process_reviews_in_batches(text_array_rotten_tomatoes, source_name='Rotten Tomatoes')
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
import time
from openpyxl import Workbook
from imdb import IMDb

# Function to get IMDb ID using IMDbPY
def get_imdb_id(title_name):
    ia = IMDb()
    try:
        search_results = ia.search_movie(title_name)
        if search_results:
            # Assuming the first result is the correct one
            movie = search_results[0]
            ia.update(movie)
            return movie.movieID
        else:
            return None
    except Exception as e:
        print(f"Error fetching IMDb ID: {e}")
        return None

# Function to get IMDb rating using IMDbPY
def get_imdb_rating(imdb_id):
    ia = IMDb()
    try:
        movie = ia.get_movie(imdb_id)
        rating = movie['rating']
        return rating
    except Exception as e:
        print(f"Error fetching IMDb rating: {e}")
        return None

# Function to get genres using IMDbPY
def get_imdb_genres(imdb_id):
    ia = IMDb()
    try:
        movie = ia.get_movie(imdb_id)
        genres = movie.get('genres', [])
        return genres
    except Exception as e:
        print(f"Error fetching IMDb genres: {e}")
        return []

# Function to scrape IMDb reviews
def scrape_imdb_reviews(title_id, max_reviews=150, max_retries=5):
    reviews = []
    pagination_key = None
    retry_attempts = 0
    while len(reviews) < max_reviews:
        try:
            if pagination_key:
                url = f"https://www.imdb.com/title/tt{title_id}/reviews/_ajax?paginationKey={pagination_key}"
            else:
                url = f"https://www.imdb.com/title/tt{title_id}/reviews?ref_=tt_ql_3"

            response = requests.get(url, timeout=10)
            response.raise_for_status()

            soup = BeautifulSoup(response.content, 'html.parser')
            review_containers = soup.find_all('div', class_='text show-more__control')
            if not review_containers:
                break  # No more reviews found

            for container in review_containers:
                if len(reviews) < max_reviews:
                    reviews.append(container.get_text())
                else:
                    break

            load_more_data = soup.find('div', class_='load-more-data')
            if load_more_data and 'data-key' in load_more_data.attrs:
                pagination_key = load_more_data['data-key']
            else:
                break

            retry_attempts = 0  # Reset retry attempts after a successful request

        except (requests.exceptions.RequestException, requests.exceptions.ConnectTimeout) as e:
            retry_attempts += 1
            if retry_attempts > max_retries:
                print(f"Max retries exceeded: {e}")
                break
            else:
                wait_time = 2 ** retry_attempts  # Exponential backoff
                print(f"Retrying in {wait_time} seconds...")
                time.sleep(wait_time)

    return reviews

# Function to save reviews to Excel
from docx import Document

def save_reviews_to_excel(reviews, title_name, source):
    # Create a new Word document
    doc = Document()

    # Add a title to the document
    #doc.add_heading(f'Reviews for: {title_name}', level=1)

    # Add source information
    #doc.add_paragraph(f'Source: {source}\n')

    # Add reviews to the document
    for idx, review in enumerate(reviews, 1):
        doc.add_paragraph(f'Review {idx}: {review}')

    # Save the document
    file_name = f"{title_name.replace(' ', ' ')}_{source}_reviews.docx"
    doc.save(file_name)
    print(f"Reviews saved to Word file: {file_name}")


# Function to find content URL on Rotten Tomatoes
def find_content_url(base_url, search_term):
    search_endpoint = '/search'
    params = {'search': search_term}
    response = requests.get(urljoin(base_url, search_endpoint), params=params)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        search_results = soup.find_all('a', class_='unset', attrs={'data-qa': 'info-name'})
        for result in search_results:
            if search_term.lower() in result.text.strip().lower():
                return urljoin(base_url, result['href'])
        print(f"No exact content link found for '{search_term}'")
        return None
    else:
        print(f'Failed to retrieve search results. Status code: {response.status_code}')
        return None

# Function to get ratings from Rotten Tomatoes
def get_ratings(content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        ratings = {}
        media_scorecard = soup.find('div', class_='media-scorecard')
        if media_scorecard:
            critics_score_button = media_scorecard.find('rt-button', attrs={
                "data-mediascorecardmanager": "overlayOpen:click",
                "slot": "criticsScore",
                "theme": "transparent"
            })
            if critics_score_button:
                critics_rt_text = critics_score_button.find('rt-text', attrs={"context": "label", "size": "1.375"})
                if critics_rt_text:
                    ratings['Critics Score'] = critics_rt_text.text.strip()
            audience_score_button = media_scorecard.find('rt-button', attrs={
                "data-mediascorecardmanager": "overlayOpen:click",
                "slot": "audienceScore",
                "theme": "transparent"
            })
            if audience_score_button:
                audience_rt_text = audience_score_button.find('rt-text', attrs={"context": "label", "size": "1.375"})
                if audience_rt_text:
                    ratings['Audience Score'] = audience_rt_text.text.strip()
        return ratings if ratings else {"Ratings": "Not found"}
    return {"Failed to retrieve content"}

# Function to get genres from Rotten Tomatoes
def get_genres(content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        genres = []
        genre_elements = soup.find_all('rt-text', attrs={"slot": "genre"})
        if genre_elements:
            genres = [genre.text.strip() for genre in genre_elements]
        if not genres:
            genre_elements = soup.find_all(class_=lambda x: x and 'genre' in x.lower())
            genres = [genre.text.strip() for genre in genre_elements]
        if not genres:
            genre_text = soup.find(text=lambda text: text and "Genre:" in text)
            if genre_text:
                genres = genre_text.find_next(text=True).strip().split(',')
        genres = list(set(filter(None, genres)))
        return genres if genres else ["Genres not found"]
    return ["Genres not found"]

# Function to navigate to reviews page on Rotten Tomatoes
def navigate_to_reviews(base_url, content_url):
    response = requests.get(content_url)
    if response.status_code == 200:
        content_soup = BeautifulSoup(response.content, 'html.parser')
        view_all_reviews_link = content_soup.find('rt-link', href=True, context='label', size='0.875')
        if view_all_reviews_link:
            reviews_url = urljoin(base_url, view_all_reviews_link['href'])
            return reviews_url
        else:
            print('View all reviews link not found on content page.')
            return None
    else:
        print(f'Failed to retrieve content page. Status code: {response.status_code}')
        return None

# Function to collect reviews using Selenium
def collect_reviews_selenium(base_url, reviews_url, max_reviews=100):
    options = webdriver.ChromeOptions()
    options.add_argument('--start-maximized')
    options.add_argument('--disable-blink-features=AutomationControlled')
    options.add_argument('--headless') # Run Chrome in headless mode
    options.add_argument('--no-sandbox') # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage') # Overcome limited resource problems
    driver = webdriver.Chrome(options=options) # Use the correct path to ChromeDriver

    # Print Chrome and ChromeDriver versions
    print("Chrome version:", driver.capabilities['browserVersion'])
    print("ChromeDriver version:", driver.capabilities['chrome']['chromedriverVersion'].split(' ')[0])

    driver.execute_cdp_cmd('Network.setUserAgentOverride', {"userAgent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
    driver.get(reviews_url)
    reviews_collected = set()
    try:
        while len(reviews_collected) < max_reviews:
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'p.review-text')))
            review_elements = driver.find_elements(By.CSS_SELECTOR, 'p.review-text')
            for element in review_elements:
                review_text = element.text.strip()
                reviews_collected.add(review_text)
            if len(reviews_collected) >= max_reviews:
                break
            try:
                load_more_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'rt-button[data-qa="load-more-btn"]')))
                driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
                time.sleep(1)
                try:
                    load_more_button.click()
                except ElementClickInterceptedException:
                    ActionChains(driver).move_to_element(load_more_button).click().perform()
                except:
                    driver.execute_script("arguments[0].click();", load_more_button)
                time.sleep(3)
                new_review_count = len(driver.find_elements(By.CSS_SELECTOR, 'p.review-text'))
                if new_review_count <= len(reviews_collected):
                    print("No new reviews loaded. Stopping.")
                    break
            except TimeoutException:
                print("Load More button not found or not clickable. Stopping.")
                break
            except Exception as e:
                print(f"Error clicking Load More button: {e}")
                break
    except Exception as e:
        print(f'Error collecting reviews: {e}')
    finally:
        driver.quit()
    reviews_list = list(reviews_collected)
    return reviews_list[:max_reviews]

# Main function
search_term = None
def main():
    base_url = 'https://www.rottentomatoes.com'
    global search_term

    # Rotten Tomatoes
    content_url = find_content_url(base_url, search_term)
    if content_url:
        print("FOR ROTTEN TOMATOES")
        #print(f"  content URL for '{search_term}' on Rotten Tomatoes: {content_url}")
        ratings = get_ratings(content_url)
        genres = get_genres(content_url)
        print("    Rotten Tomatoes Ratings:")
        if 'Audience Score' in ratings:
            print(f"      Audience Rating: {ratings['Audience Score']}")
        if 'Critics Score' in ratings:
            print(f"      Critics Score: {ratings['Critics Score']}")
        if 'Audience Rating' not in ratings and 'Critics Score' not in ratings:
            print("Ratings not found")
        print(f"      Rotten Tomatoes Genres: {', '.join(genres)}")
        reviews_url = navigate_to_reviews(base_url, content_url)
        if reviews_url:
            #print(f"Reviews URL on Rotten Tomatoes: {reviews_url}")
            collected_reviews = collect_reviews_selenium(base_url, reviews_url, max_reviews=200)
            #print(f"Total collected reviews for '{search_term}' on Rotten Tomatoes: {len(collected_reviews)}")
            save_reviews_to_excel(collected_reviews, search_term, "RottenTomatoes")
        else:
            print(f"Failed to retrieve reviews URL for '{search_term}' on Rotten Tomatoes")
    else:
        print(f"No exact content link found for '{search_term}' on Rotten Tomatoes")

    # IMDb
    imdb_id = get_imdb_id(search_term)
    if imdb_id:
        print("FOR IMDB")
        print(f"IMDb ID for '{search_term}': {imdb_id}")

        # Fetch IMDb rating
        rating = get_imdb_rating(imdb_id)
        if rating:
            print(f"  IMDb Rating: {rating}")
        else:
            print(f"No IMDb rating found for '{search_term}'")

        # Fetch genres
        genres = get_imdb_genres(imdb_id)
        if genres:
            print(f"  Genres: {', '.join(genres)}")
        else:
            print(f"No genres found for '{search_term}'")

        # Scrape IMDb reviews
        print(f"\nFetching reviews for '{search_term}' from IMDb DONE")
        imdb_reviews = scrape_imdb_reviews(imdb_id, max_reviews=100)  # Adjust max_reviews as needed
        if imdb_reviews:
            #print("\nSaving reviews to Excel...")
            save_reviews_to_excel(imdb_reviews, search_term, "IMDb")
        else:
            print("No IMDb reviews found.")
    else:
        print(f"No IMDb ID found for '{search_term}'")




if __name__ == "__main__":
    global search_term
    search_term = input("Enter the name of the movie, series, or show: ")
    main()
```

```python
import time
import re
from docx import Document
from gradio_client import Client

# Function to extract text from a Word document
def extract_text_from_word(file_path):
    doc = Document(file_path)
    return [paragraph.text for paragraph in doc.paragraphs if paragraph.text.strip()]

# Load reviews from IMDb Word document

imdb_source_file = f"/content/{search_term}_IMDb_reviews.docx"
text_array_imdb = extract_text_from_word(imdb_source_file)

# Load reviews from Rotten Tomatoes Word document
rotten_tomatoes_source_file = f"/content/{search_term}_RottenTomatoes_reviews.docx"
text_array_rotten_tomatoes = extract_text_from_word(rotten_tomatoes_source_file)

# Initialize Gradio client
client = Client("Nymbo/Llama-3.1-405B-Instruct-FP8")

# Function to process reviews in batches using Gradio client
def process_reviews_in_batches(reviews, batch_size=8, source_name='Reviews'):
    num_reviews = len(reviews)
    mean_scores = []

    for i in range(0, num_reviews, batch_size):


        # Create the message for the current batch
        user_message = f"Just give the average sentiment score for reviews ranging from -1 to 1, 1 being the best and -1 being the worst. It can be continuous values like 0.5, 0.9, -0.5. {reviews[i:i + batch_size]}. Please don't give the scores for individual reviews and also don't give any extra information."

        # Generate the sentiment score using the Gradio client
        result=client.predict(
		message=user_message,
		system_message="You are a friendly Chatbot.",
		max_tokens=512,
		temperature=0.7,
		top_p=0.95,
		api_name="/chat"
)

        # Use regex to extract the mean score from the model's output
        match = re.search(r'[-+]?\d*\.\d+|\d+', result.strip())
        if match:
            mean_score = float(match.group(0))  # Convert the captured string to float
            mean_scores.append(mean_score)

            # Print the mean score for the current batch
            print(f"Mean Sentiment Score for {source_name} reviews {i+1} to {min(i + batch_size, num_reviews)}: {mean_score}")
        else:
            print(f"Failed to extract mean score from the model's output: '{result.strip()}'")

        # Wait for 20 seconds before processing the next batch
        time.sleep(20)

    # Calculate the overall mean score across all batches
    overall_mean_score = sum(mean_scores) / len(mean_scores) if mean_scores else 0
    print(f"Overall Mean Sentiment Score for {source_name}: {overall_mean_score}")

# Process the IMDb reviews in batches
process_reviews_in_batches(text_array_imdb, source_name='IMDb')

print("##--##--##--##--##--##--##--##--##")

# Process the Rotten Tomatoes reviews in batches
process_reviews_in_batches(text_array_rotten_tomatoes, source_name='Rotten Tomatoes') # Commented out for now
```
